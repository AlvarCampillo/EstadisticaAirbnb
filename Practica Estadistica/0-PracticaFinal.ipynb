Cual es el precio medio de alquiler de cada uno, la diferencia que hay ¿es estadísticamente significativa?
¿Con que test lo comprobarías?

```r
# Cálculo de medias
paste("La media de precios de Entire Home/Apartment es: ", mean(airbnb[airbnb$TipoAlquiler == "Entire home/apt", 12], na.rm = TRUE))
paste("La media de precios de Private room es: ", mean(airbnb[airbnb$TipoAlquiler == "Private room", 12], na.rm = TRUE))
paste("La media de precios de Shared Room es: ", mean(airbnb[airbnb$TipoAlquiler == "Shared room", 12], na.rm = TRUE))

# Principios de Normalidad e Igualdad de varianzas
Entero<-airbnb[airbnb$TipoAlquiler == "Entire home/apt", 12]
SubEntero<- sample(Entero, 5000, replace=FALSE)
shapiro.test(SubEntero)

Privada<-airbnb[airbnb$TipoAlquiler == "Private room", 12]
SubPrivada<- sample(Privada, 5000, replace=FALSE)
shapiro.test(SubPrivada)

Compartida<-airbnb[airbnb$TipoAlquiler == "Shared room", 12]
shapiro.test(Compartida)

var.test(Entero,Privada)
var.test(Entero,Compartida)
var.test(Privada,Compartida)

# Test Comparación de medias y post-hoc
print("Los test de Shapiro-Wilk muestran que no se cumple la hipótesis de normalidad, pero sí la de homocedasticidad, por lo que no se utilizará un ANOVA si no el test de Kruskal-Wallis")

kruskal.test( Precio ~ TipoAlquiler, data=airbnb)

print("Como se ve en los resultados del test de Kruskal-Wallis, sí existen diferencias significativas entre al menos dos de los grupos")

pairwise.wilcox.test(x = airbnb$Precio, g = airbnb$TipoAlquiler, p.adjust.method = "holm" )

print("Utilizando una prueba post-hoc, se puede ver que hay diferencias significativas entre todos los grupos")
```

Esta claro que las medias de metros cuadrados de cada uno de estos 5 barrios parecen ser diferentes, pero ¿son estadísticamente diferentes?
¿Que test habría que usar para comprobarlo?

```r
#usamos shapiro test para ver si las distribuciones son normales
NRecoletos<-airbnb_entire[airbnb_entire$Barrio == "Recoletos",14]
NGoya<-airbnb_entire[airbnb_entire$Barrio == "Goya",14]
NTrafalgar<-airbnb_entire[airbnb_entire$Barrio == "Trafalgar",14]
NSol<-airbnb_entire[airbnb_entire$Barrio == "Sol",14]
NJusticia<-airbnb_entire[airbnb_entire$Barrio == "Justicia",14]

shapiro.test(NRecoletos)
shapiro.test(NGoya)
shapiro.test(NTrafalgar)
shapiro.test(NSol)
shapiro.test(NJusticia)

print("Se tratan de distribuciones no normales a excepción de 1, por ello habrá que usar la prueba de Kruskal-Wallis")
kruskal.test(list(NRecoletos,NGoya,NTrafalgar,NSol,NJusticia))

print("Al ser el p-value superior a lo establecido (0,05), podemos afirmar que no existen diferencias significativas con respecto a los metros cuadrados entre los pisos de estos 5 barrios")
```
